```python
df1sortedpergameweek=df1.sort_values('gameweek') # i sorted the dataset per gameweek in order to get a list with the cumulative
# sum of the point for each team during all the league. after that i can plot it
```

```python
def keysteams():
    lst=[]
    lstkey=[]
    for i in range(380):
        lst+=list(df1sortedpergameweek['teamsData'][i].keys())
    for k in lst:
        if k not in lstkey:
            lstkey+=[k]
    return(lstkey)
keysteams()
# this function returns me a list with all the id of the team
```

```python
teamnames=['Burnley','Bournemouth','Crystal Palace','WBA','Arsenal','Huddersfield','Brighton','Liverpool','Watford','Man. Utd','Newcastle','Chelsea','Man. City','Southhampton','Swansea','Stoke City','Leicester','Tottenham','Everton','West Ham']
def teamid():
    diz={}
    for i in range(20):
        diz[keysteams()[i]]=teamnames[i]
    return diz
teamid()
# this function returns a dictionary where the keys are the id of the team while the values are the names of the teams
```

```python
def calendarperteam(idteam):
    match=[]
    points=[]
    for k in range(380):
        if idteam in list(df1sortedpergameweek['teamsData'][379-k].keys()):
            match+=[df1sortedpergameweek['winner'][379-k]]
    for j in match:
        if int(idteam)==j:
            points+=[3]
        elif int(idteam)!=j and j!=0:
            points+=[0]
        elif int(idteam)!=j and j==0:
            points+=[1]
    return(points)
calendarperteam('1625')
# it returns the points get by the team (input: id of the team) for each match of the league from the first gameweek to the last
```

```python
cumsumpoint=[list(np.cumsum(list(calendarperteam(i)))) for i in keysteams()]
cumsumpoint
# this function apply to all the teams the cumulative sum of the point during the league. now i can plot it
```

```python
def conswin(idteam):
    counter=1
    maxl=[]
    for i in range(1,38):
        if calendarperteam(idteam)[i]==3 and calendarperteam(idteam)[i-1]==3:
            counter+=1
        elif calendarperteam(idteam)[i]==3 and calendarperteam(idteam)[i-1]!=3 or calendarperteam(idteam)[i]!=3:
            counter=1
        maxl+=[counter]
    if max(maxl)==1:
        maxl=[]
        maxl.append(0)
    return([max(maxl),teamid()[idteam]])
conswin('1625')
# this function takes in input the idteam and gives the longest streak of win with the name of the team
```

```python
def conslose(idteam):
    counter=1
    maxl=[]
    for i in range(1,38):
        if calendarperteam(idteam)[i]==0 and calendarperteam(idteam)[i-1]==0:
            counter+=1
        elif calendarperteam(idteam)[i]==0 and calendarperteam(idteam)[i-1]!=0 or calendarperteam(idteam)[i]!=0:
            counter=1
        maxl+=[counter]
    if max(maxl)==1:
        maxl=[]
        maxl.append(0)
    return([max(maxl),teamid()[idteam]])
conslose('1609')
# the same function of the previous one but it returns the longest streak of defeats with the name of the team
```

```python
consecutivewinperteam=[conswin(i) for i in keysteams()]
consecutiveloseperteam=[conslose(i) for i in keysteams()]
# i apply the last function to all the teams
```

```python
function splitSwap(a, l, n):
  if n <= 1:
    return
  splitSwap(a, l, n/2)
  splitSwap(a, l+ n /2, n/2)
  swapList(a, l, n)
The subroutine swapList is described here:

function swapList(a, l, n):
  for i = 0 to n/2:
    tmp = a[l + i]
    a[l + i] = a[l + n/2 + i]
    a[l + n/2 + i] = tmp
```

##Ex: l=[a,b,c,d,e,f,g,h]

step1: [[a,b,c,d] , [e,f,g,h]]

step2: [[e,f,g,h] , [a,b,c,d]]

step3: [[[e,f] , [g,h]] , [[a,b] , [c,d]]]

step4: [[[g,h] , [e,f]] , [[c,d] , [a,b]]]

step5: [[[[g] , [h]] , [[e] , [f]]] , [[[c] , [d]] , [[a] , [b]]]]

step6: [[[[h] , [g]] , [[f] , [e]]] , [[[d] ,[c]] , [[b] , [a]]]]

step7: [h,g,f,e,d,c,b,a]

The main goal of this algoritm is to reverse the input array. The idea is to devide the input into 2 subs and change their place. The same function will apply on the subs until the lenght of them equals to 1. Then we have a reverse sequense of elements and array is reversed.

T(n) = O(nlog n)
